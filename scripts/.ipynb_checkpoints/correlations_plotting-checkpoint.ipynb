```python
def save_paired_genes_to_csv(grouped, clearance_filename, risk_filename, chunk_size=10):
    """
    Saves clearance genes and associated risk genes to CSV files, aligning rows so that
    each clearance gene is repeated for each chunk of risk genes.

    :param grouped: Grouped pandas Series with clearance genes as the index and associated risk genes as values.
    :param clearance_filename: Name of the CSV file to save clearance genes.
    :param risk_filename: Name of the CSV file to save risk genes.
    :param chunk_size: Maximum number of risk genes per line in the output file.
    """
    with open(clearance_filename, 'w') as clearance_file, open(risk_filename, 'w') as risk_file:
        jobs = 0
        for clearance_gene, risk_genes in grouped.items():
            jobs += math.ceil(len(risk_genes) / chunk_size)
            for i in range(0, len(risk_genes), chunk_size):
                # Write clearance gene to its file
                formatted_clearance = f'["{clearance_gene}"]'
                clearance_file.write(formatted_clearance + '\n')
                
                # Write the corresponding chunk of risk genes to its file
                chunk = risk_genes[i:i + chunk_size]
                formatted_chunk = '["' + '", "'.join(chunk) + '"]'
                risk_file.write(formatted_chunk + '\n')

    print(f'Wrote {jobs} jobs to {clearance_filename} and {risk_filename}')
```

```python
def decode_paired_genes_from_csv(clearance_filename, risk_filename):
    """
    Decodes paired clearance and risk genes from two CSV files into a single DataFrame.

    :param clearance_filename: Name of the CSV file containing clearance genes.
    :param risk_filename: Name of the CSV file containing risk genes.
    :return: DataFrame with clearance genes and their associated risk genes.
    """
    clearance_genes = []
    risk_genes = []

    with open(clearance_filename, 'r') as clearance_file, open(risk_filename, 'r') as risk_file:
        for clearance_line, risk_line in zip(clearance_file, risk_file):
            clearance_gene = clearance_line.strip().strip('["]').split('", "')[0]
            risk_gene_list = risk_line.strip().strip('["]').split('", "')
            clearance_genes.extend([clearance_gene] * len(risk_gene_list))
            risk_genes.extend(risk_gene_list)

    return pd.DataFrame({'clearance gene': clearance_genes, 'risk gene': risk_genes})
```

```python
# removes bad lines from csv file, saves clean file, and outputs missing rows

label_dir = '/home/gchan/kg98_scratch/gchan/SIR_SCZ/SIR_simulator/SIR_matlab/data/gene_labels/'
my_path = '/home/gchan/kg98_scratch/gchan/SIR_SCZ/SIR_results/lme_beta/S132_spatial.csv'
my_df = pd.read_csv(my_path, on_bad_lines='skip')
# remove rows where correlation values have been written in to gene names
my_df = my_df[my_df.iloc[:, :1].apply(lambda x: x.str.contains('^[A-Z]', regex=True)).all(axis=1)]
# remove rows where gene names have been written in to correlation matrix
my_df.iloc[:, 2:] = my_df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')
my_df = my_df.dropna()
my_df.to_csv(my_path, index=False)

# generates all expected gene pair combinations
if ('rewired' in my_path) | ('spatial' in my_path):
    all_combinations = decode_paired_genes_from_csv(label_dir + 'lme_clear.csv', label_dir + 'lme_risk.csv')
else:
    all_combinations = pd.MultiIndex.from_product([my_df['clearance gene'].unique(), my_df['risk gene'].unique()], names=['clearance gene', 'risk gene']).to_frame(index=False)

# finds missing combinations and saves them to csv files
merged_df = all_combinations.merge(my_df, on=['clearance gene', 'risk gene'], how='left', indicator=True)
missing_rows = merged_df[merged_df['_merge'] == 'left_only'].drop(merged_df.columns[2:], axis=1)
grouped = missing_rows.groupby('clearance gene')['risk gene'].apply(list)
save_paired_genes_to_csv(grouped, label_dir + 'rerun_clearance_genes.csv', label_dir + 'rerun_risk_genes.csv', chunk_size=1000)
print('Number of missing rows:', len(missing_rows))
print('Number of clearance genes:', len(grouped))
```

```python
# 4 min for 4M rows
from scipy.stats import zscore, norm
my_df=gene_df[(gene_df['mean_deviation'] < thresh) & (gene_df['correlation'] > 0.5)]

# for null_df, null_type in zip([rewired_df, spatial_df], ['rewired', 'spatial']):
for null_df, null_type in zip([spatial_df], ['spatial']):
    # Merge the null model dataframes with the original dataframe
    merge_df = my_df.merge(null_df, how='left', left_index=True, right_index=True, suffixes=('', '_' + null_type))
    # Calculate z-scores based on the summary statistics
    merge_df['z_' + null_type] = (merge_df['correlation'] - merge_df['mean']) / merge_df['SD']
    # Convert z-scores to p-values
    merge_df['p_' + null_type] = merge_df['z_' + null_type].apply(lambda z: 2 * (1 - norm.cdf(abs(z))))
    merge_df = merge_df.sort_values(by='correlation', ascending=False).reset_index(drop=True)

    # Display the updated dataframe
    display(merge_df.head())

out_df = merge_df[['clearance gene', 'risk gene', 'correlation', 'mean_deviation', 'p_spatial']]
out_df.to_csv('/home/gchan/kg98_scratch/gchan/SIR_SCZ/SIR_results/lme_beta/S132_corrs_p-vals.csv', index=False)

num_significant_pairs = ((merge_df['p_spatial'] < 0.05) & (merge_df['correlation'] > 0.5)).sum()
print(f'Number of pairs with p-value < 0.05: {num_significant_pairs}')
```
